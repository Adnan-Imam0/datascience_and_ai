# What is MatplotLib

### Matplotlib is a foundational python library for creating static, interactive and animated plots. It allows for detail customization of plots

### basic Syntax

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [10,20,40,40,50]
plt.plot(x,y)
plt.show()
```

## Basic Plots

### Line plot

```python
# line plot is used to visualoze trends over time or sequence
```

```python
plt.plot(x,y, label="Trends")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot")
plt.legend()
plt.show()
```

### Bar Chart

```python
# bar charts is used to compare categorical data
```

```python
categories = ["A", "B", "C"]
values = [10, 20, 30]
plt.bar(categories, values, color = "Blue")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart")
plt.show()
```

## Histogram

```python
# Histogram is used to show the distribution of Data in a dataset
```

```python
data = [1,2,3,2,1,3,4,2,4,5,4,5,2,1]
plt.hist(data, bins = 5, color = "Green", edgecolor = "Black")
plt.xlabel("Data")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

### Scatter Plot

```python
# A Scatter plot visualizes the relationship between two continuous variavles and used quite a lot in Ai projects
```

```python
plt.scatter(x,y, color = "Red", marker = "o")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.show()
```

### Customization Of Plots

#### we can add titles, axis labels, legends, Adjust colors and styles

```python
plt.plot(x,y, color = "orange", marker = "x", linestyle= "--")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(["dataset 1"])
plt.title("Line Plot")
plt.show()
```

### 1. Create a basic Plots with matplotlib

```python
# line plot
years = [2010,2011,2012,2013]
sales = [100, 80, 120, 200]
plt.plot(years, sales, marker = "o", linestyle = "--", label= "sales Trend")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.legend()
plt.title("Sales Over years")
plt.show()
```

```python
# bar Chart
categories = ["electronics", "Clothings", "Groceries"]
revenue = [100, 270, 160]
plt.bar(categories, revenue, color = "green", label = "Revenue")
plt.legend()
plt.xlabel("Categories")
plt.ylabel("Revenue")
plt.title("Revenue Analysis by Category")
plt.show()
```

```python
# Scatter plot
hour_studied = [1,2,2,3,5,4]
exam_score = [50,60,59, 65,80,72]
plt.scatter(hour_studied, exam_score, color= "green")
plt.xlabel("Hour Studied")
plt.ylabel("Exam Score")
plt.title("Study Hours vs Exam Scores")
plt.show()
```

#### EX.2. create a basic HeatMap with Seaborn

```python
# the objective here is to create heatmap from a correlation matrix
# instructions that i can follow here is first i load a data set and calculate correlation matrix
# then i am going to visualize the correlation matrix using a heatmap
```

```python
import pandas as pd
```

```python
df = sns.load_dataset("iris")
```

```python
df.head()
```

```python
# calculate the correlation matrix
# we are using only numeric columns
correlation_matrix = df.corr(numeric_only=True)
```

```python
correlation_matrix
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
